```python
import pandas as pd
```

```python
data = pd.read_csv("/content/drive/MyDrive/PrepinstaPrime/Excel Datasets/Copy of Data-cleaning-for-beginners-using-pandas.csv")
data
```

```python
data["Age"] = data["Age"].fillna(round(data["Age"].mean(), 1))
data["Age"] = data["Age"].astype("int");
data
```

```python
data["Salary"] = data["Salary"].str.replace('k','000')
data["Salary"] = data["Salary"].str.replace('$','')
data
```

```python
data["Rating"] = data["Rating"].apply(lambda x: None if x == -1 else x)
data
```

```python
# Location
data['Location'] = data['Location'].replace({'India,In': 'India', 'New York,Ny':'New York','Australia Aus': 'Australia', 'India In': 'India', 'Austraila Aus': 'Australia'})
data
```

```python
data["Established"] = data["Established"].apply(lambda x: None if x == -1 else x)
data
```

```python
data["Easy Apply"] = data["Easy Apply"].str.replace("-1", "FALSE")
data
```
